# 柱狀圖（Bar Chart）

# 圓餅圖（Pie Chart）

適合呈現比例（資料的結果與分佈）



````python
import matplotlib.pyplot as plt

# 產生資料
labels = ["Python", "Java", "C++", "Javascript"]
sizes = [30, 27, 23, 20]

# 畫出圖表
plt.pie(sizes, labels=labels)
plt.axis('equal')
plt.show()
````

````python
import matplotlib.pyplot as plt
 
# 產生資料
labels = ["Python", "Java", "C++", "Javascript"]
sizes = [30, 27, 23, 20]

# 畫出圖表
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
# 等比例
plt.axis('equal')
plt.show()
````

````python
import matplotlib.pyplot as plt
 
# 產生資料
labels = ["Python", "Java", "C++", "Javascript"]
sizes = [30, 27, 23, 20]
colors = ['blue', 'red', 'orange', 'green']
# 畫出圖表
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
````

````python
import matplotlib.pyplot as plt
 
# 產生資料
labels = ["Python", "Java", "C++", "Javascript"]
sizes = [30, 27, 23, 20]
colors = ['blue', 'red', 'orange', 'green']
explode = (0.1, 0, 0, 0)  # 將 Python 的部分從圓餅圖分開
 
# 畫出圖表
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True)
 
plt.axis('equal')
plt.show()
````

# 增加圖例 (Legend)

````python
import matplotlib.pyplot as plt
 
# 產生資料
labels = ["Python", "Java", "C++", "Javascript"]
sizes = [30, 27, 23, 20]
colors = ['blue', 'red', 'orange', 'green']
explode = (0.1, 0, 0, 0)  # 將 Python 的部分從圓餅圖分開
 
# 畫出圖表
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True)

# 設定圖例
plt.legend(loc='right')
plt.axis('equal')
plt.show()
````

# 範例：畫出歷史股價走勢圖

````python
import numpy as np
from datetime import datetime

# 產生一個長度為 31,數值在 200 至 250 之間的亂數陣列
prices = np.random.uniform(200, 250, 31)
prices
````

````python
import pandas as pd

# 產生一個日期序列
time_idx = pd.date_range(start="2018/12/1", end="2018/12/31")
time_idx
````

````python
plt.plot(time_idx, prices)
plt.xlabel("日期")
plt.ylabel("價格")
plt.title("歷史股價走勢圖")
plt.show()
````

# Matplotlib 的中文顯示問題

由於 Matplotlib 顯示中文有問題，針對這個問題，有多種不同的解法，其中最簡單的解法是，需要將 Matplotlib 中參數字 `rcParams` 中字體的無襯綫字體屬性 `font.sans-serif` 改成 `SimHei` (等寬體) 或是 `Microsoft JhengHei` (微軟正黑體)

請打開字體資料夾，安裝 `SimHei` 或是 `Microsoft JhengHei` 字體

````python
# 設定字體
plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei"]
````

````python
plt.plot(time_idx, prices)
plt.xlabel("日期")
plt.ylabel("價格")
plt.title("歷史股價走勢圖")
plt.show()
````

若上面的方法依然沒有解決顯示中文的問題，可以試試 `Matplotlib` 的 `FontProperties`

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
# 指定字體的檔案位置與大小
font = FontProperties(fname=r"C:\到 msjf.ttf 檔案的路徑", size=14) 
# 接下來就可以用 fontproperties 屬性設定字體
plt.xlabel("x軸", fontproperties=font)
plt.ylabel("y軸", fontproperties=font)
plt.title("標題", fontproperties=font)
plt.show()
```

上述方法在語法上較爲冗長，但卻是相對保險的做法，更多解決方案請參考：[連結](https://codertw.com/%E7%A8%8B%E5%BC%8F%E8%AA%9E%E8%A8%80/359974/)

````python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

plt.plot(time_idx, prices)
font = FontProperties(
    fname=r"字體//msjh.ttf", 
    size=14) # 指定字體的檔案位置與大小
plt.xlabel("日期", fontproperties=font) # 指定每一個 label 會使用的字體
plt.ylabel("價格", fontproperties=font)
plt.title("歷史股價走勢圖", fontproperties=font)
plt.show()
````

````python
plt.plot(time_idx, prices)
plt.xlabel('日期', fontproperties=font)
plt.ylabel('價格', fontproperties=font)
plt.title('歷史股價走勢圖', fontproperties=font)
# plt.grid 設定圖表的網格
plt.grid(True, axis='y')
plt.show()
````

````python
import numpy as np
from datetime import datetime
# 產生一個長度為 31,數值在 200 至 250 之間的亂數陣列
prices2 = np.random.uniform(230, 280, 31)
# 產生一個日期序列
time_idx = pd.date_range(start='2018/12/1', end='2018/12/31')
# 將兩個綫條顯示在同一張圖表上
plt.plot(time_idx, prices2)
plt.plot(time_idx, prices)
plt.xlabel("日期", fontproperties=font)
plt.ylabel("價格", fontproperties=font)
plt.title("歷史股價走勢圖", fontproperties=font)
plt.grid(True, axis='y')
plt.show()
````

## 線條的樣式

在用 Matplotlib 畫圖時，線條的樣式可以用指定 linestyle 來表示：

| 型態 | 名稱 | 符號 |
|:--:|:------:|:-----:|
| 實綫 | 'solid' | '-' |
| 虛點 | 'dashed' | '--' |
| 綫點 | 'dashdot' | '-.' |
| 點綫 | 'dotted' | ':' |


````python
plt.plot(time_idx, prices2, label="股票1", linestyle="--")
plt.plot(time_idx, prices, label="股票2", ls="dashdot")
plt.xlabel("日期", fontproperties=font)
plt.ylabel("價格", fontproperties=font)
plt.title("歷史股價走勢圖", fontproperties=font)
plt.legend()
plt.grid(True, axis='y')
plt.show()
````

綫條的顔色可以指定 color 來表示：

| 顔色 | 簡寫 | 
|:--:|:------:|
| 'red' | 'r' |
| 'black' | 'k' |
| 'blue' | 'b' | 
| 'cyan' | 'c' | 
| 'yellow' | 'y' | 
| 'white' | 'w' | 
| 'green' | 'g' | 

````python
plt.plot(time_idx, prices2, label="股票1", linestyle="--", color="red")
plt.plot(time_idx, prices, label="股票2", ls="dashdot", c="b")
plt.xlabel("日期", fontproperties=font)
plt.ylabel("價格", fontproperties=font)
plt.title("歷史股價走勢圖", fontproperties=font)
plt.legend()
plt.grid(True, axis='y')
plt.show()
````

點綫的樣式則可透過 marker 設定：

| 形狀 | 英文 | 符號 | 
|:--:|:------:|:------:|
| 點 | point | '.' |
| 圓圈 | circle | 'o' |
| 三角形往下 | triangle_down | 'v' | 
| 三角形往上 | triangle_up | '^' | 
| 三角形往左 | triangle_left | '<' | 
| 三角形往右 | triangle_right | '>' | 
| 正方形 | squared | 's' | 
| 加號 | plus | '+' |
| 星號 | star | '*' |

````python
plt.plot(time_idx, prices2, label="股票1", linestyle="--", color="red", marker="o")
plt.plot(time_idx, prices, label="股票2", ls="dashdot", c="b", marker="*")
plt.xlabel("日期", fontproperties=font)
plt.ylabel("價格", fontproperties=font)
plt.title("歷史股價走勢圖", fontproperties=font)
plt.legend()
plt.grid(True, axis='y')
plt.show()
````

````python
df = pd.DataFrame(prices, index=time_idx, columns=['價格'])
# 到這邊我們也注意到了，原來 DataFrame 的 plot 方法，其實就是 matplotlib 的 pyplot
fig = df.plot(grid=True, legend=True, marker="o")
fig.set_xlabel("日期（日）", fontproperties=font)
fig.set_ylabel("價格（NTD）", fontproperties=font)
fig.set_title("歷史股價走勢圖", fontproperties=font)
plt.show()
````
