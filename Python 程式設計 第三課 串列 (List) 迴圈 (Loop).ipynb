# 迴圈（Loop）

就是把同樣的程式碼重複執行很多次 (迭代)

# 迴圈
```python

avengers = ["Ironman", "Captain America", "Hulk", "Black Widow", "Thor", "Hawkeye"]

# 接下來我們就把 avengers 内的每一筆資料，依照順序遍歷一次
# 每迭代到一筆資料，就放入變數 i
for i in avengers:
	# 每次迭代時，變數 i 的值就會被 print() 顯示
    print(i)

# Ironman
# Captain America
# Hulk
# Black Widow
# Thor
# Hawkeye
```

````python
avengers = ["Ironman", "Captain America", "Hulk", "Black Widow", "Thor", "Hawkeye"]
for i in avengers:
	# 每次迭代時，變數 i 的值就會被 print() 顯示
    print(i)
````

# 簡單來說...

就是把一段**同樣的程式碼重複執行很多次 (迭代)**

但是每一次的迭代又會把清單裡的其中一筆資料放入變數

而開發者可以利用這個變數在**每一次的迭代算出不一樣的結果**


# 迴圈的結構(1)

- 宣告迴圈的種類（for）
- 宣告一個 iterator（avenger）
- 宣告一個迴圈要跑的iterable（avengers）
- 縮排宣告每一次 iterator 變動的時候要做什麼事（print(avenger)）


# 迴圈的結構(2)

```python
for iterator in iterable:
    # 用放入 iterator 的值去做一些事...
```

```python

```

# break, continue, pass 語句

很多時候迴圈都會結合判斷式，以達到程式碼執行的靈活度

# break

停止執行迴圈

```python
lst = ["a", "b", "c", "d"]

for i in lst:
    print(i)
    if i == "c":
        break # 當迭代到 "c" 時，跳出迴圈，"d" 不會被印出
```

````python
lst = ["a", "b", "c", "d"]
for i in lst:
    print(i)
    if i == "c":
        break # 當迭代到 "c" 時，跳出迴圈，"d" 不會被印出
````

# continue

跳過不執行某一個迭代，直接執行下一個迭代

```python
lst = ["a", "b", "c", "d"]

for i in lst:
    if i == "c":
        continue # 當迭代到 "c" 時，跳過 c，直接迭代下一個元素
    print(i)
```

````python
lst = ["a", "b", "c", "d"]

for i in lst:
    if i == "c":
        continue # 當迭代到 "c" 時，跳過 c，直接迭代下一個元素
    print(i)
````

# pass

占位語句，爲了避免一些尚未寫完整的程式碼在執行時出錯：

例:
```python
lst = [1,2,3,4,5]

for i in lst:
# 這樣寫會出錯，python 不允許 for 迴圈底下沒有未縮排的程式碼
```

所以我們就在 for 迴圈底下加上一個 **pass**:

```python
lst = [1,2,3,4,5]

for i in lst:
    pass
```

則可以被執行

````python

lst = [1,2,3,4,5]

for i in lst:
    pass
````


# 回家作業 1

計算 1 到 100 的**標準差**

[標準差 Wiki](https://zh.wikipedia.org/wiki/%E6%A8%99%E6%BA%96%E5%B7%AE)

```python
# 產生出一個 1 到 100 的 list
data = list(range(1, 101))
# 算出標準差
```

1. 用 for 迴圈實作
2. 用 while 迴圈實作

````python
import statistics
#用 for 迴圈實作
data=list(range(1,101))
mean=statistics.mean(data)
output=0
for i in data:
    output +=  (i - mean)**2
SD=(output/(len(data)-1))**0.5
print(SD)
#用 while 迴圈實作
data=1
mean=statistics.mean(range(1,101))
output=0
while data<=100:
    output +=  (data - mean)**2
    data += 1
SD=(output/(data-2))**0.5
print(SD)
````

# 回家作業 2

把之前的剪刀、石頭、布遊戲加上以下功能：

1. 檢查使用者出拳是否是在選擇範圍內 (1,2,3)
2. 詢問使用者是否再玩一次，若使用者回答是肯定的，就再玩一次


````python
import random
print("歡迎來到剪刀石頭布！ ")
#name = input("請輸入您的名稱：")
win_solution = [[1,3], [2,1], [3,2]]
print("請出拳： (1) 剪刀 (2) 石頭 (3) 布：")
player_hand=0;time=1;win=0;lose=0;tie=0
for time in range(1,4):
    while True:
        player_hand=input()
        if player_hand in ['1','2','3']:
            player_hand=int(player_hand)
            break
        else:print('請輸入1或2或3')
    print(f'第{time}把')
    time +=1
    computer_hand = random.randint(1,3)
    output=['剪刀','石頭','布']
    #print(f"你贏{win}次 輸{lose}次 平手{tie}次")
    print(f"你出{output[player_hand-1]}")
    print(f"電腦出{output[computer_hand-1]}")
    if computer_hand==player_hand:
        tie +=1
        print('平手')        
    elif [computer_hand,player_hand] in win_solution:
        lose +=1
        print('電腦贏')
    else:
        print('你贏')
        win +=1
print(f"你贏{win}次，輸{lose}次，平手{tie}次")
````
